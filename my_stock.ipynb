```python
from urllib.request import urlopen
import pandas as pd
import plotly.graph_objects as go
```

```python
## Samsung Electronics
url = 'https://query1.finance.yahoo.com/v7/finance/download/005930.KS?period1=1576559279&period2=1608181679&interval=1d&events=history&includeAdjustedClose=true'
```

```python
## LG Chem
# url = "https://query1.finance.yahoo.com/v7/finance/download/051910.KS?period1=1576584445&period2=1608206845&interval=1d&events=history&includeAdjustedClose=true"
```

```python
sources = urlopen(url).readlines()
sources[-5:]
```

```python
columns = [x.decode('utf-8').rstrip('\n').split(',') for x in sources[:1]]
columns
```

```python
columns = columns[0]
columns
```

```python
deco_source = [x.decode('utf-8').rstrip('\n').split(',') for x in sources[1:]]
deco_source[:5]
```

```python
df = pd.DataFrame(deco_source, columns = columns)
df.head()
```

```python
df.Date.tolist()[1]
```

```python

trace1 = go.Scatter(
        x=df.Date.tolist(),
        y=[float(y) for y in df.High],
        name = "High",
        line = dict(color="#e74c3c"),
        opacity=0.8)

trace2 = go.Scatter(
        x=df.Date.tolist(),
        y=[float(y) for y in df.Low],
        name = "Low",
        line = dict(color="#3498db"),
        opacity=0.8)

trace3 = go.Scatter(
        x=df.Date.tolist(),
        y=[float(y) for y in df.High],
        name = "High",
        line = dict(color="#e74c3c"),
        opacity=0.8)

frames = [go.Frame(
        data=[go.Scatter(
        x=[df.Date.tolist()[k]],
        y=[[float(y) for y in df.High][k]],
        name = "m_High",
        mode="markers",
        marker=dict(color="yellow", size=20, opacity=1))])
        for k in range(len(df.Date))]


data = [trace1, trace2, trace3]

frame = frames

layout = dict(title="Stock Test",
              margin = dict(l=0, r=0, b=0, t=0),
              updatemenus=[dict(type="buttons",buttons = [dict(
                        args = [None, {"frame": {"duration": 20, "redraw": False},
                                       "fromcurrent": True, "transition": {"duration": 0}}],
                                        label = "Play",
                                        method = "animate")])])
```

```python
fig = go.Figure(data=data, frames=frame, layout=layout)
fig.update_xaxes(rangeslider_visible=True)

fig.show()
fig.write_html("./results/test8_stock.html")
```
