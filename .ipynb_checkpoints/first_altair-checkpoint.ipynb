```python
import numpy as np
import pandas as pd
from pandas_datareader import data
import altair as alt
from datetime import datetime
```

```python
now = datetime.now()
now
```

```python
dt_string = now.strftime("%Y-%m-%d")
type(dt_string)
```

```python
start = '2019-1-1'
end = dt_string
source = 'yahoo'
apple = data.DataReader("AAPL", start=start ,end=end, data_source=source).reset_index()
ibm = data.DataReader("IBM", start=start ,end=end, data_source=source).reset_index()
microsoft = data.DataReader("MSFT", start=start ,end=end, data_source=source).reset_index()
```

```python
apple['Symbol'] = 'AAPL'
ibm['Symbol'] = 'IBM'
microsoft['Symbol'] = 'MSFT'
```

```python
apple.tail()
```

```python
stocks = pd.concat(
    [apple[['Date','Close','Volume','Symbol']],
     ibm[['Date','Close','Volume','Symbol']],
     microsoft[['Date','Close','Volume','Symbol']]],
     axis=0)
```

```python
stocks.head()
```

```python
(alt.
  Chart(stocks[stocks.Symbol == 'AAPL']).
  mark_line().
  encode(x='Date', y='Close'))
```

```python
(alt.
  Chart(stocks).
  mark_line().
  encode(x='Date', y='Close', color='Symbol').
  properties(height=300, width=500))
```

```python
start = '2019-1-1'
end = dt_string
source = 'yahoo'
```

```python
samsung = data.DataReader("005930.KS", start=start ,end=end, data_source=source).reset_index()
hynix = data.DataReader("000660.KS", start=start ,end=end, data_source=source).reset_index()
# LG_Chem = data.DataReader("051910.KS", start=start ,end=end, data_source=source).reset_index()
```

```python
samsung['Symbol'] = 'samsung'
hynix['Symbol'] = 'hynix'
```

```python
stocks1 = pd.concat(
    [samsung[['Date','Close','Volume','Symbol']],
     hynix[['Date','Close','Volume','Symbol']]],
     axis=0)
```

```python
(alt.
  Chart(stocks1).
  mark_line().
  encode(x='Date', y='Close', color='Symbol').
  properties(height=300, width=500))
```

```python
input_dropdown = alt.binding_select(options=['samsung','hynix'])
# selection = alt.selection_single(fields=['Symbol'], bind=input_dropdown, name='Close1')
scales = alt.selection_interval(bind='scales')   #   

line = (alt.
         Chart(stocks1).
         mark_line().
         encode(x='Date', y='Close', color='Symbol', tooltip='Close').
         properties(height=400, width=500).
         add_selection(scales).
         transform_filter(scales))


avg = (alt.
         Chart(stocks1).
         mark_rule().
         encode(y='average(Close)', color='Symbol',   
                size=alt.value(2)))

line + avg
```
