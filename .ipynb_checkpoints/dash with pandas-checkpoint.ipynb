```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.graph_objs as go
import numpy as np
import pandas as pd
```

```python
app = dash.Dash()
```

```python
src = pd.read_csv('./source/WHO-COVID-19-global-data.csv')
src
```

```python
df = src.loc[src.Country == 'Japan']
df
```

```python
app.layout = html.Div([
    
    dcc.Graph(
        id = 'scatter_chart',
        figure = {
            'data':[
                go.Scatter(
                    x = df.Date_reported,
                    y = df.New_cases,
                    mode='markers'
                )
            ],
            'layout':go.Layout(
                title="Covid19 Visualization",
                xaxis={'title':'Date'},
                yaxis={'title':'New Cases'}
            )
        }

    )
])
```

```python
if __name__ == '__main__':
    app.run_server(port = 4023)
```
