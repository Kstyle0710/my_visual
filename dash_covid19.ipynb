```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.graph_objs as go
import numpy as np
import pandas as pd
from datetime import datetime as dt
```

```python
app = dash.Dash()
```

```python
src = pd.read_csv('./source/WHO-COVID-19-global-data.csv')
src
```

```python
df = src.loc[src.Country == 'Republic of Korea']
df
```

```python
app.layout = html.Div([
    html.Label('Choose a Country'),
    dcc.Dropdown(
    id = 'first-dropdown',
    options = [
        {'label': 'Korea', 'value':'Republic of Korea'},
        {'label': 'Japan', 'value':'Japan'},
        {'label': 'China', 'value':'China', 'disabled':True} # 개별 항목 비활성화
        ],
        #value = "Republic of Korea",      # dafault value
        #multi = True,   # 복수 항목 선택 가능
        #placeholder = "Select a Country",    # 플레이스 홀더 표시
        #disabled = True    # 드랍다운 박스 비활성화
    ),
    html.Br(),
    html.Br(),
    
    html.Label("This is a Slide"),
    dcc.Slider(  ## 양방향 슬라이더
        min = 1,
        max = 10,
        value = 5,     # 초기 값
        step = .5,     # 슬라이드 포인트의 이동 간격
        marks = {i:i for i in range(10)}
    ),
    html.Br(),
    html.Br(),
    
    html.Label("This is a range slider"),
    dcc.RangeSlider(
        min = 1, 
        max = 10,
        step = .5,
        value = [3, 7],  # 초기 값 (레인지는 범위이기 때문에 시작과 끝 두개의 값)
        marks = {i:i for i in range(10)}     
    ),
    html.Br(),
    html.Br(),
    
    html.Label("This is a Input Box"),
    dcc.Input(
        placeholder = "Input your value",
        type = 'text',
        value = ""
    
    ),
    
    html.Br(),
    html.Br(),
    html.Label("This is Button"),
    html.Button("Submit", id='submit-form'),
    
    html.Br(),
    html.Br(),
    html.Label("This is a Text Area"),
    html.Br(),
    dcc.Textarea(
        placeholder = "write your long feedback",
        value = 'text',
        style = {'width': '100%'}      
    ),
    
    html.Br(),
    html.Br(),
    html.Label("This is Check Box"),
    dcc.Checklist(
        options = [
            {'label': 'Korea', 'value':'Republic of Korea'},
            {'label': 'Japan', 'value':'Japan'},
            {'label': 'China', 'value':'China', 'disabled':True} # 개별 항목 비활성화
            ],
            value = ['Republic of Korea']     # dafault value
    ),
    
    html.Br(),
    html.Br(),
    html.Label("This is Radio Button"),
    dcc.RadioItems(
        options = [
            {'label': 'Korea', 'value':'Republic of Korea'},
            {'label': 'Japan', 'value':'Japan'},
            {'label': 'China', 'value':'China', 'disabled':True} # 개별 항목 비활성화
            ],
            value = 'Republic of Korea'    
    ),
    
    html.Br(),
    html.Br(),
    html.Label("This is Date Picker Single"),
    dcc.DatePickerSingle(
        id = 'df-pick-single',
        date = dt(2020, 7, 10)
    ),
    
    html.Br(),
    html.Br(),
    html.Label("This is Date Picker Range"),
    dcc.DatePickerRange(
        id = 'df-pick-range',
        start_date = dt(2020, 7, 10),
        end_date = 
    ),

    
])
```
