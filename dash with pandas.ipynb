```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.graph_objs as go
import numpy as np
import pandas as pd
```

```python
app = dash.Dash()
```

```python
src = pd.read_csv('./source/WHO-COVID-19-global-data.csv')
src
```

```python
df = src.loc[src.Country == 'Republic of Korea']
df
```

```python
app.layout = html.Div([
    
    dcc.Graph(
        id = 'scatter_chart',
        figure = {
            'data':[
                go.Line(
                    x = df.Date_reported,
                    y = df.New_cases,
                    mode='line'
                )
            ],
            'layout':go.Layout(
                title="Covid19 Visualization",
                xaxis={'title':'Date'},
                yaxis={'title':'New Cases'},
                hovermode='closest'  # 커서가 근처에 가야 호버 표시 (이게 없으면 수직 라인만 맞으면 멀리 있어도 호버 표시)
            )
        }

    )
])
```

```python
if __name__ == '__main__':
    app.run_server(debug=False)
```
